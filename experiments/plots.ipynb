```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def accuracy_plot(keys, values):
    fig = plt.figure()
    ax = fig.add_axes([0,0,1,1])
    ax.set_ylabel('Accuracy (in percent)')
    ax.set_yticks(np.arange(0, 100, 5))
    ax.set_title('Accuracy per genre')
    ax.axhline(np.array(list(values)).mean(), color='blue', linewidth=2)
    plt.xticks(rotation = 45)
    plt.ylim([0,100])
    plt.bar(keys, values)
```

```python
accuracies = {"Rock": 68,
              "Folk": 74,
              "International": 47,
              "Hip-Hop": 72,
              "Pop": 15,
              "Electronic": 46,
              "Experimental": 13,
              "Instrumental": 33, }

accuracy_plot(accuracies.keys(), accuracies.values())
```
