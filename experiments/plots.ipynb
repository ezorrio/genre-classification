```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def accuracy_plot(title, keys, values):
    fig = plt.figure()
    ax = fig.add_axes([0,0,1,1])
    ax.set_title(title)
    ax.set_ylabel('Accuracy (in percent)')
    ax.set_yticks(np.arange(0, 100, 10))
    ax.set_title('Accuracy per genre')
    ax.axhline(np.array(list(values)).mean(), color='blue', linewidth=2)
    plt.xticks(rotation = 45)
    plt.ylim([0,100])
    plt.bar(keys, values)
    plt.savefig(f"{title}.png", dpi=150)
```

```python
#training vs validation
accuracies = {"Rock": 70,
              "Folk": 66,
              "International": 52,
              "Hip-Hop": 61,
              "Pop": 15,
              "Electronic": 43,
              "Experimental": 22,
              "Instrumental": 31, }

accuracy_plot("Training vs validation", accuracies.keys(), accuracies.values())
```

```python
#training vs test
accuracies = {"Rock": 50,
              "Folk": 14,
              "International": 38,
              "Hip-Hop": 58,
              "Pop": 21,
              "Electronic": 27,
              "Experimental": 9,
              "Instrumental": 40, }

accuracy_plot("Training vs test", accuracies.keys(), accuracies.values())
```

```python
#training+validation vs test
accuracies = {"Rock": 48,
              "Folk": 8,
              "International": 46,
              "Hip-Hop": 59,
              "Pop": 23,
              "Electronic": 32,
              "Experimental": 8,
              "Instrumental": 47, }

accuracy_plot("Training and validation vs test", accuracies.keys(), accuracies.values())
```
